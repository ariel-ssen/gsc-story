```python
#doGrade 내가 만들어준 이름 
def doGrade (d) :
    if d >= 30 :
        return 'A'
    elif d >= 20 :
        return 'B'
    else :
        return 'C'
```

```python
#Grade  내가 만들어준 이름
mpg['Grade'] = mpg['total'].apply(doGrade)
```

```python
mpg['Grade'].value_counts()
```

```python
#value_counts 를 index 순서대로 정렬하고 싶을때 sort_index()
mpg['Grade'].value_counts().sort_index()
```

```python
data[['total','asian']]
```

```python
#3
data['percent'] =( data['asian'] / data['total']) *100
```

```python
data['percent']
```

```python
#4 >large | < small
```

```python
data['percent'].mean()
```

```python
data['size'] = np.where(data['percent'] > data['percent'].mean(),'large','small')
```

```python
data['size']
```

```python
data['size'].value_counts()
```

---

```python
def isSize(percent, mean) :
    if percent > mean :
        return 'large'
    return 'small'
```

```python
data['size2'] = data['percent'].apply(isSize, args=[data['percent'].mean()])
```

```python
data['size2'].value_counts()
```
