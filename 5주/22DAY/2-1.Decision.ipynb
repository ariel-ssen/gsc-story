- 엔트로피 공식

```python
# class_probabilities = [p, q]
class_probabilities = [5/10, 5/10]
```

```python
sum(-p * math.log(p, 2) for p in class_probabilities if p)
```

- 엔트로피 구하는 함수 
- 1에 가까우면 혼잡 0 에 가까우면 안정

```python
def entropy(class_probabilities) :
    return sum(-p * math.log(p, 2) for p in class_probabilities if p)
    # if p를 해주는 이유는 None값이 들어갔을때 오류가 나기 때문에 해주는게 좋음
```

- 추가설명

```python
entropy(class_probabilities)
```

```python
def class_probabilities(labels) : 
    total = len(labels)
    return [count / total for count in Counter(labels).values()]
```

```python
def data_entropy(labeled_data) :
    labels = [label for _, label in labeled_data]
    #print('labels : ', labels)
    # 언더바 되어 있는 부분은 출력하지말고, label에 해당하는 값만 출력해라.
    probabilities = class_probabilities(labels)
    #print('probabilities : ', probabilities)
    return entropy(probabilities)
```
